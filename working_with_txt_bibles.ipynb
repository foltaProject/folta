```
class TxtBible:
  def __init__(self, filename):
    bible_file = open(filename)
    self.language = bible_file.readline()[24:].strip()
    self.iso = bible_file.readline()[24:].strip()
    self.iso15924 = bible_file.readline()[24:].strip()
    self.year = bible_file.readline()[24:].strip()
    self.years = bible_file.readline()[24:].strip()
    self.title_vernacular = bible_file.readline()[24:].strip()
    self.title_english = bible_file.readline()[24:].strip()
    self.url = bible_file.readline()[24:].strip()
    self.short_copy = bible_file.readline()[24:].strip()
    self.long_copy = bible_file.readline()[24:].strip()
    self.notes = bible_file.readline()[24:].strip()
    verses = bible_file.readlines()
    self.verse_ids = [x[:8] for x in verses]
    self.verse_txt = [x[9:].strip() for x in verses]
    tokens = [y for x in self.verse_txt for y in x.split()]
    self.verse_ct = len(self.verse_ids)
    self.token_ct = len(tokens)
    self.vocab = set(tokens)
    self.n_vocab = len(self.vocab)

```

This Class can be used to process the .txt-formatted bibles on Awakateko.

`self.language` - calls for the language of the bible, spelled out.

* Example: `English`

`self.iso` - calls for the 3-digit ISO code of the language.

* Example: `eng`

`self.iso15924` - calls what I believe to be the script used

* Example: `Latn`

`self.year` - calls for the latest date of copyright

* Example: `2011`

`self.years` - calls for all of the copyright dates

* Example: `1979; 1984; 2011`

`self.title_vernacular` - calls for the name of the bible in the language of the text

* Example: `Holy Bible, New International Version® Anglicised`

`self.title_english` - calls for the name of the text in English

* Example: `The Bible in English`

`self.url` - calls for the url to the text

* Example: `https://www.bible.com/bible/113/gen.1.nivuk`

`self.short_copy` - calls for the short form of the copyright

* Example: `© Biblica, Inc.® 2011`

`self.long_copy` - calls for the long form of the copyright

* Example: `Holy Bible, New International Version®, Anglicised, NIV®<br>Copyright © 1979, 1984, 2011 by Biblica, Inc.® Used by permission. All rights reserved worldwide.<br>The Holy Bible, New International Version®, Anglicised, NIV® is copyrighted and has been made available on the internet for your personal use only. Any other use including, but not limited to, copying or reposting on the internet is prohibited. These Scriptures may not be altered or modified in any form and must remain in their original context. These Scriptures may not be sold or otherwise offered for sale.<br>These Scriptures are not shareware and may not be duplicated. These scriptures are not public domain.`

`self.notes` - text from the 'notes' field in the metadata

* Example: (field empty in English test bible)

`self.verse_ids` - returns a list of verse labels

* Example: `['40001001', '40001017', '40001018',...`
* Note: ID format = 'BBCCCVVV'
1. The first two digits of the index represent the book designation.
 * Example: `1` = `Genesis` (First book of bible
 * Example: `40` = `Matthew` (First book of New Testament.
 * Example: `67` = `Tobit` (First book of apocryphal writings)
2. The next three digits represent the  chapter.
3. The final three digits represent the verse.

`self.verse_txt` - list of verse text strings, correspondent to verse_IDs

* Example: `['This is the genealogy of...`

`self.verse_ct` - number of verses in the text

* Example: `7713`

`self.token_ct` - number of total tokens in the text

* Example: `200326`

`self.vocab` - set of unique vocabulary

* Example: `{'disturbed', 'uncircumcision', ...`

`self.n_vocab` - size of vocabulary

* Example: `7103`
